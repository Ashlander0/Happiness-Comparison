```python
import pandas as pd

# input year for mastering
year = '2015'

# filepath for origin happiness file
originfilepath = f'Datasets/{year}.csv'

# export tocation for master file
exportpath = f'Clean_Datasets/{year}_master.csv'
```

```python
# import happiness CSV
happiness = pd.read_csv(originfilepath)

# year is 2018 or 2019 (country column named different
if year == '2018' or year == '2019':
    happiness = happiness.rename(columns={'Country or region': 'Country'})



# import CSVs to be added
mil = pd.read_csv('Clean_Datasets/Mil_Percent_GDP.csv')
popgrowth = pd.read_csv('Clean_Datasets/Population_Growth.csv')
pdensity = pd.read_csv('Datasets/population_density_per_square_km.csv')
refugee_from = pd.read_csv('Datasets/refugee_from_perc_wb.csv')
refugee_in = pd.read_csv('Datasets/refugee_in_perc_wb.csv')

happiness
```

```python
# seperate added data by year selected
mil = mil[['Country', year]]
popgrowth = popgrowth[['Country', year]]
pdensity = pdensity[['country', year]].rename(columns={'country': 'Country'})
refugee_from = refugee_from[['country', year]].rename(columns={'country': 'Country'})
refugee_in = refugee_in[['country', year]].rename(columns={'country': 'Country'})

mil
```

```python
# create and add new data to master DF
master = pd.merge(happiness, mil, on='Country')                         # merge with military expenditure
master = master.rename(columns={year: 'Military Expenditure of GDP'})
master = pd.merge(master, popgrowth, on='Country')                      # merge with population growth
master = master.rename(columns={year: 'Population Growth'})
master = pd.merge(master, pdensity, on='Country')                       # merge with population density
master = master.rename(columns={year: 'Population Density'})
master = pd.merge(master, refugee_from, on='Country')                   # merge with refugees from
master = master.rename(columns={year: 'Pop % as Refugees Abroad'})
master = pd.merge(master, refugee_in, on='Country')                     # merge with refugees in
master = master.rename(columns={year: 'Pop % as Refugee Intake'})

master
```

```python
# delete any NaN
master = master.dropna(axis=0, how='any')

master
```

```python
# export to master file

master.to_csv(exportpath)
```
